## Regression

```julia
function plotMEG(surprises, megs, ols)
    fig = figure(figsize=(3,3.5))
    
    # set title and labels
    xlabel("Surprise (bits)")
    ylabel("MEG activity (pT)")
    grid(true)
    
    scale = 1e12
    
    # plot points
    for i = eachindex(surprises)
        scatter(surprises[i], scale * megs[i], s=1, label="block $i")
    end
    
    # plot regression line
    coefs = coef(ols)
    points = map(length, megs)
    intercept = sum(coefs[1:end-1] .* points) / sum(points)
    
    X = reduce(vcat, surprises)
    yhat = coefs[end] * X .+ intercept
    
    r2val = @sprintf("%.3f %%", 100 * r2(ols))
    plot(X, scale * yhat, "k", label="linear regression, r2 = $r2val")

    # legend()
end;
```

## Maximum $R^2$

Compute the maximum possible $R^2$ of linear regressions for different possible combinations of window lengths and parameters. The results are stored in an external file such as to be reusable and allow for interruptions during the execution of the cell.

```julia
function export_maxr2(seq, filename; dir = "ckpt")
    f = open("$dir/$filename", "w")

    for el in seq
        print(f, el)
        print(f, " ")
    end
    
    close(f)
end

function import_maxr2(filename; dir = "ckpt")
    f = open("$dir/$filename", "r")
    t = readdlm(f)
    close(f)
    
    return t[1,1:321]
end
;
```

The function `mapssh` maps the data from a subject to Shannon surprise values from the learning rule given as parameter.

```julia
?mapssh
```

```julia
function computeR2(X::Array{Float64,2}, y::Array{Float64,1})
    return r2(fit(LinearModel, X, y))
end

function mapmaxr2(subject, models, subjectidx = 1, onehotencode_blocks = false, verbose = 2; dir = "ckpt")   
    ssh, megs = mapssh(subject, models, onehotencode_blocks, verbose)
    
    # ssh is (n_models, n_rules_per_model, surprises) (in array form)
    # we need to cat so we get (n_models x n_rules_per_model, surprises)
    ssh = vcat(ssh...)
    
    verbose > 1 && lg("mapmaxr2: computing r^2...")
    
    # map to r2
    r2s = zeros(size(ssh,1), size(megs,2), size(megs,3))
    for i in CartesianIndices(r2s)
        r2s[i] = computeR2(ssh[i[1]], megs[:,i[2],i[3]])
    end
    
    # max for each sensor across all learning rules
    r2s = @pipe findmax(r2s, dims=1) |> dropdims(_[1], dims=1)
    
    # average across sensors
    r2s = @pipe mean(r2s, dims=1) |> dropdims(_, dims=1)
    
    # save to ckpt folder
    filename = "max_r2_subject_$subjectidx.txt"
    export_maxr2(r2s, filename, dir=dir)
    
    return r2s
end
;
```

```julia
function mapmaxr2_across_subjects(subjects, models, onehotencode_blocks = false, verbose = 1; dir = "ckpt")
    timesteps = 321
    maxr2 = zeros(length(subjects), timesteps)
    t = range(-250, stop=1000, length=timesteps) |> collect
    
    for subidx in eachindex(subjects)
        subname = subjects[subidx]
        verbose > 0 && lg("Loading data for subject $subname...")
        
        # compute r2 timecourse
        filename = "max_r2_subject_$subname.txt"
        if isfile("$dir/$filename")
            maxr2[subidx,:] = import_maxr2(filename, dir=dir)
        else
            subdata = load_jld2(get_jld2_filename(subname))
            maxr2[subidx,:] = mapmaxr2(subdata, models, subname, onehotencode_blocks, verbose, dir=dir)
        end
    end
    
    # average across subjects
    maxr2_avg = @pipe mean(maxr2, dims=1) |> dropdims(_, dims=1)
    
    return t, maxr2_avg, maxr2
end
;
```

```julia
function plot_best_r2(t, maxr2_avg, maxr2)
    N = size(maxr2,1) # number of subjects
    maxr2_ste = @pipe std(maxr2, dims = 1) |> dropdims(_, dims=1) ./ sqrt(N)
    maxr2_min = maxr2_avg - maxr2_ste
    maxr2_max = maxr2_avg + maxr2_ste
    
    figure(figsize=[10,5])
    
    # plot average line
    plot(t, 100 * maxr2_avg, "k", lw=2)
    # plot standard error
    fill_between(t, 100 * maxr2_min, 100 * maxr2_max, color="0.75", alpha=0.5)
    
    xlabel("Time (ms)")
    ylabel("Explained variance (%)")
    grid("on")
end
;
```

### Leaky integration, $m \in [0, 1]$

```julia
# set range of window lengths
window_lengths = [0,1]

# set range of leak parameters
ws = vcat(log2range(0, 10, 54), [Inf])

# leaky integration
leaky_models = map(
    m -> map(w -> Leaky(w, prior_fixed(m), updateallcols=true), ws),
    window_lengths
)

# subjects
subjects = 1:18

# keep block level?
onehotencode_blocks = false;
```

```julia
# get max r2
verbose = 0
dir = "ckpt/r2/leaky_noblock_01"
t, maxR2_, maxR2List = mapmaxr2_across_subjects(subjects, leaky_models, onehotencode_blocks, verbose, dir=dir);
```

```julia
plot_best_r2(t, maxR2_, maxR2List)

# add red patches at 3 time-windows
ax = gca()
slices = [[60, 130], [160, 320], [460, 625]]

for slice in slices
    x = slice[1]
    w = slice[2] - slice[1]
    rect = matplotlib.patches.Rectangle((x, 0), w, 1, linewidth=1, facecolor=(1,0,0,0.1))
    ax.add_patch(rect)
end
```

### Leaky integration, $m \in [0, 1, 2]$

```julia
leaky_models = map(
    m -> map(w -> Leaky(w, prior_fixed(m), updateallcols=true), ws),
    [0,1,2]
);
```

```julia
# get max r2
verbose = 0
dir = "ckpt/r2/leaky_noblock_012"
t, maxR2_, maxR2List = mapmaxr2_across_subjects(subjects, leaky_models, onehotencode_blocks, verbose, dir=dir);
```

```julia
plot_best_r2(t, maxR2_, maxR2List);
```

### Variational SMiLe

```julia
pcs = 1 .- exp.(- 1 ./ ws)
ms = pcs ./ (1 .- pcs)

varsmile_models = map(
    wlen -> map(m -> VarSMiLe(m, prior_fixed(wlen), updateallcols=true), ms),
    [0,1,2]
);
```

```julia
# get max r2
verbose = 0
dir = "ckpt/r2/varsmile_noblock_01"
t, maxR2_, maxR2List = mapmaxr2_across_subjects(subjects, leaky_models, onehotencode_blocks, verbose, dir=dir);
```

```julia
plot_best_r2(t, maxR2_, maxR2List);
```
