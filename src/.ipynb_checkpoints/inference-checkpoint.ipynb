```julia
using PyPlot, Printf

include("decode.jl")
include("utils.jl")
include("inference.jl")

include("UpdateRule.jl")
include("Leaky.jl")
include("VarSMiLe.jl")
include("ParticleFiltering.jl")
;
```

```julia
function decode_posterior_with_inference(seq::Array{<:Integer,1},
                                         rule::UpdateRule,
                                         prior::Array{Float64,1},
                                         N = 2)
    # define callback to compute sbf and theta
    function callback(r, x_t, col)
        return r.params()[2]
    end
    cb = Callback(callback, Array{Float64,1})

    # decode sequence
    return decode_with_inference(seq, rule, prior, cb, N)
end
;
```

```julia
function plot_heatmap(values, xticks, xlabels, yticks, ylabels, title, filename, xlabel)
    heatmap = imshow(values', origin="lower left", cmap="rainbow", vmin=0, vmax=1)
    ax = gca()

    # set x-axis
    ax.set_xlabel(xlabel)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xlabels)

    # set y-axis
    ax.set_ylabel(L"p")
    ax.set_yticks(yticks)
    ax.set_yticklabels(ylabels)

    # colorbar at the end
    cbar = colorbar(heatmap)

    # add title
    PyPlot.title(title)
    display(gcf())
    
    savefig(filename)
    
    clf()
end

function plot_heatmap_w(values, xticks, xlabels, yticks, ylabels, title, filename)
    return plot_heatmap(values, xticks, xlabels, yticks, ylabels, title, filename,
        L"Timescale of integration ($\omega$)")
end

function plot_heatmap_p(values, xticks, xlabels, yticks, ylabels, title, filename)
    return plot_heatmap(values, xticks, xlabels, yticks, ylabels, title, filename,
        L"Reciprocal of change probability ($\frac{1}{p_c}$)")
end
;
```

```julia
# sequence length
L = 300

# heatmap params
resolution = 21
ps = range(0, 1, length=resolution) |> collect

# prior for models
prior = prior_uniform(1)
;
```

## Leaky integration

```julia
# heatmap params
ws = vcat(log2range(0, 10, resolution-1), Inf);
```

```julia
leaky_avgs  = zeros(length(ws), length(ps))
leaky_freqs = zeros(length(ws), length(ps))

leaky_runs = 0;
```

```julia
for k in 1:30
    for (i, w) in enumerate(ws), (j, p) in enumerate(ps)
        # generate random sequence of length L, with P(x = 1) = p
        seq = rand(Bernoulli(p), L)

        # create update rule    
        rule = LeakyInference(w, true, true)

        # decode sequence
        posterior = decode_posterior_with_inference(seq, rule, prior)

        # store as result
        #leaky_mins[i,j]  = findmin(posterior[:,2])[1]
        #leaky_maxs[i,j]  = findmax(posterior[:,2])[1]
        leaky_avgs[i,j]  += mean(posterior)[2]
        leaky_freqs[i,j] += mean(map(p -> p[2] > 0.5, posterior))
    end
    leaky_runs += 1
end
;
```

```julia
step = round(Int, (resolution - 1) / 10)
xticks = vcat(0:step:resolution-1)
xlabels = map(i -> ws[i+1] == Inf ? L"\infty" : @sprintf("%i", ws[i+1]), xticks)

yticks = collect(0:step:resolution-1)
ylabels = map(i -> @sprintf("%.1f", ps[i+1]), yticks)

plot_heatmap_w(leaky_avgs  / leaky_runs,  xticks, xlabels, yticks, ylabels, 
    L"Average value of $P(m = 1)$", "leaky_average.svg")
plot_heatmap_w(leaky_freqs / leaky_runs, xticks, xlabels, yticks, ylabels, 
    L"Frequency of $P(m = 1) > \frac{1}{2}$", "leaky_freq.svg")
;
```

## Particle Filtering

```julia
# heatmap params
pcs = 1 .- exp.(- 1 ./ ws)
ms = pcs ./ (1 .- pcs);
```

```julia
par_avgs  = zeros(length(ms), length(ps))
par_freqs = zeros(length(ms), length(ps))

Nparticles = 100
Nthr = Nparticles / 2

par_runs = 0;
```

```julia
# results
for k in 1:30
    for (i, m) in enumerate(ms), (j, p) in enumerate(ps)
        # generate random sequence of length L, with P(x = 1) = p
        seq = rand(Bernoulli(p), L)

        # create update rule    
        rule = ParticleFilteringInference(m, Nparticles, Nthr, true)

        # decode sequence
        posterior = decode_posterior_with_inference(seq, rule, prior)

        # store as result
        #par_mins[i,j]  = findmin(posterior[:,2])[1]
        #par_maxs[i,j]  = findmax(posterior[:,2])[1]
        par_avgs[i,j]  += mean(posterior)[2]
        par_freqs[i,j] += mean(map(p -> p[2] > 0.5, posterior))
    end
    par_runs += 1
end
;
```

```julia
step = round(Int, (resolution - 1) / 10)
xticks = vcat(0:step:resolution-1)
xlabels = map(i -> ms[i+1] == 0 ? L"\infty" : @sprintf("%i", 1/ms[i+1] + 1), xticks)

yticks = collect(0:step:resolution-1)
ylabels = map(i -> @sprintf("%.1f", ps[i+1]), yticks)

plot_heatmap_p(par_avgs / par_runs,  xticks, xlabels, yticks, ylabels, 
    L"Average value for $P(m = 1)$", "particle_filtering_average.svg")
plot_heatmap_p(par_freqs / par_runs, xticks, xlabels, yticks, ylabels,
    L"Frequency of $P(m = 1) > \frac{1}{2}$", "particle_filtering_freq.svg")
;
```

## Variational SMiLE

```julia
# results
#var_mins  = Array{Float64,2}(undef, length(ms), length(ps))
#var_maxs  = Array{Float64,2}(undef, length(ms), length(ps))
var_avgs  = Array{Float64,2}(undef, length(ms), length(ps))
var_freqs = Array{Float64,2}(undef, length(ms), length(ps))

for (i, m) in enumerate(ms), (j, p) in enumerate(ps)
    # generate random sequence of length L, with P(x = 1) = p
    seq = rand(Bernoulli(p), L)
    
    # create update rule    
    rule = VarSMiLeInference(m, true)
    
    # decode sequence
    posterior = decode_posterior_with_inference(seq, rule, prior) |> unwrap
    
    # store as result
    #var_mins[i,j]  = findmin(posterior[:,2])[1]
    #var_maxs[i,j]  = findmax(posterior[:,2])[1]
    var_avgs[i,j]  = mean(posterior[:,2])
    var_freqs[i,j] = mean(posterior[:,2] .> 0.5)
end
;
```

```julia
step = round(Int, (resolution - 1) / 10)
xticks = vcat(0:step:resolution-1)
xlabels = map(i -> ms[i+1] == 0 ? L"\infty" : @sprintf("%i", 1/ms[i+1] + 1), xticks)

yticks = collect(0:step:resolution-1)
ylabels = map(i -> @sprintf("%.1f", ps[i+1]), yticks)

#plot_heatmap_p(var_mins,  xticks, xlabels, yticks, ylabels, L"Min value for $P(m = 1)$")
#plot_heatmap_p(var_maxs,  xticks, xlabels, yticks, ylabels, L"Max value for $P(m = 1)$")
plot_heatmap_p(var_avgs,  xticks, xlabels, yticks, ylabels, L"Average value for $P(m = 1)$")
plot_heatmap_p(var_freqs, xticks, xlabels, yticks, ylabels, L"Frequency of $P(m = 1) > \frac{1}{2}$")
;
```
